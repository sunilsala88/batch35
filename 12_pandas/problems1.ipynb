```python
import pandas as pd
import requests

url = 'https://www.screener.in/screens/281185/nifty-50/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
df7 = pd.read_html(response.text)
df=df7[0]
```

```python
df=df.drop(15,axis=0)
```

```python
df['P/E']=df['P/E'].astype('float')
```

```python
df.info()
```

```python
df[df['P/E']>100]['Name'].to_list()
```
