```python
import numpy as np
import pandas as pd

np1=np.arange(25).reshape(5,5)
df1=pd.DataFrame(np1,index=['r1','r2','r3','r4','r5'],columns=['a','b','c','d','e'])
df1.index.name='r'
df1
```

```python
np2=np.arange(100,125).reshape(5,5)
df2=pd.DataFrame(np2,index=['r1','r2','r3','r4','r6'],columns=['c1','c2','c3','c4','c5'])
df2.index.name='r'
df2
```

```python
#left,right,innner,outer
pd.merge(df1,df2,on='r',how='outer')
```

```python
np1=np.arange(25).reshape(5,5)
df3=pd.DataFrame(np1,columns=['r1','r2','r3','r4','r5'],index=['a','b','c','d','e'])
# df1.index.name='r'
print(df3)

np2=np.arange(100,125).reshape(5,5)
df4=pd.DataFrame(np2,columns=['r1','r2','r3','r4','r6'],index=['c1','c2','c3','c4','c5'])
# df2.index.name='r'
print(df4)
```

```python
pd.concat([df3,df4],join='inner')
```

```python
import pandas as pd
import requests

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
df7 = pd.read_html(response.text)
data2=df7[0]
data2
```

```python
data1=data1.rename(columns={'Company name':'Name','Sector[15]':'Sector','Date added[16]':'Date'})
data1
```

```python
data2=data2.rename(columns={'Security':'Name','GICS Sector':'Sector','Date added':'Date'})
data2
```
