```python
import numpy as np
import pandas as pd

np1=np.arange(25).reshape(5,5)
df1=pd.DataFrame(np1,index=['r1','r2','r3','r4','r5'],columns=['a','b','c','d','e'])
df1.index.name='r'
df1
```

```python
np2=np.arange(100,125).reshape(5,5)
df2=pd.DataFrame(np2,index=['r1','r2','r3','r4','r5'],columns=['c1','c2','c3','c4','c5'])
df2.index.name='r'
df2
```

```python
pd.merge(df1,df2,on='r')
```
