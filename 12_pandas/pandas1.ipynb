```python

import numpy as np
np1=np.arange(25).reshape(5,5)
np1
```

dataframe from numpy array

```python
import pandas as pd
df1=pd.DataFrame(np1,index=['r1','r2','r3','r4','r5'],columns=['c1','c2','c3','c4','c5'])
df1
```

dataframe from list of list

```python
l1=[[1,2,3],[44,55,66],[77,88,99]]
df2=pd.DataFrame(l1,columns=['tsla','amzn','goog'],index=[11,12,13])
df2
```

dataframe from dict of list

```python
dl1=[{"name":'tsla','price':500,'sector':'auto'},
     {"name":'amzn','price':600,'sector':'tech'},
     {"name":'jpmorgan','price':800,'sector':'bank'}]
df3=pd.DataFrame(dl1)
df3
```

list of dictionary

```python
ld1={"name":['tsla','amzn','jpmorgan'],
     'price':[500,600,800],
     'sector':['auto','tech','bank']}
df4=pd.DataFrame(ld1)
df4
```

```python
import pandas as pd
import requests

url = 'https://en.wikipedia.org/wiki/NIFTY_50'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
df7 = pd.read_html(response.text)
df7=df7[1]
```

```python
df7.rename(columns={'Company name':'name','Sector[15]':'sector','Date added[16]':'Date'},inplace=True)
```
